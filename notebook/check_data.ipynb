```python
import pandas as pd

# Path to your CSV file
csv_path = "/home/ubuntu/project/Tumor_Detection/ratio_out.csv"

# Load CSV into a DataFrame
df = pd.read_csv(csv_path)

# Select only class ratio columns (skip 'patch_name')
class_columns = ["background_ratio", "stroma_ratio", "benign_ratio", "tumor_ratio"]

# Compute statistics
stats = df[class_columns].agg(['min', 'mean', 'max'])

# Rename 'mean' to 'avg' for clarity
stats = stats.rename(index={'mean': 'avg'})

# Display results
print(stats.T.round(6))  # Transpose for better readability
```

```python
import pandas as pd

# Path to your CSV file
csv_path = "/home/ubuntu/project/Tumor_Detection/ratio_out.csv"

# Load the CSV
df = pd.read_csv(csv_path)

# Class columns (excluding patch_name)
class_columns = ["background_ratio", "stroma_ratio", "benign_ratio", "tumor_ratio"]

# Define percentiles to compute
percentiles = [0.0, 0.25, 0.5, 0.7, 0.9, 1.0]

# Compute percentiles
stats = df[class_columns].quantile(percentiles)

# Label the index nicely
stats.index = ['0%', '25%', '50%', '70%', '90%', '100%']

# Transpose for better readability (each class as row)
print(stats.round(6))
```

```python
import numpy as np
import torch 
numDataPoints = 1000
data_dim = 5
bs = 100

# Create dummy data with class imbalance 9 to 1
data = torch.FloatTensor(numDataPoints, data_dim)
target = np.hstack((np.zeros(int(numDataPoints * 0.9), dtype=np.int32),
                    np.ones(int(numDataPoints * 0.1), dtype=np.int32)))

print ('target train 0/1: {}/{}'.format(
    len(np.where(target == 0)[0]), len(np.where(target == 1)[0])))

class_sample_count = np.array(
    [len(np.where(target == t)[0]) for t in np.unique(target)])
weight = 1. / class_sample_count
samples_weight = np.array([weight[t] for t in target])

samples_weight = torch.from_numpy(samples_weight)
samples_weigth = samples_weight.double()


len(samples_weight), samples_weight
```
